```python
def get_image_coordinates(header, w):
    m = header["NAXIS1"]
    n = header["NAXIS2"]
    try:
        frame = header["RADESYS"].lower()
    except KeyError:
        frame = "fk5"
    x = np.arange(0, m, 1)
    y = np.arange(0, n, 1)
    xx, yy = np.meshgrid(x, y)

    ra, dec = w.all_pix2world(xx, yy, 0) * un.deg
    sky_coord = SkyCoord(ra=ra, dec=dec, frame=frame)

    return sky_coord
```
