```python
import astropy.units as u
import matplotlib.pyplot as plt
from astropy.io import fits

from csromer.plots import Plotter
```

```python
radio_image_name = "/home/miguel/Documents/datasets/A1314_JVLA/mfs/A1314_I.fits"
xray_image_name = "/home/miguel/Documents/datasets/A1314_JVLA/xray.fits"
```

```python
hdul_radio = fits.open(radio_image_name)[0]
hdul_xray = fits.open(xray_image_name)
```

```python
plotter = Plotter(
    radio_image=hdul_radio,
    xray_image=hdul_xray,
    use_latex=True,
    z=0.034,
    scalebar_length_kpc=100 * u.kpc
)
```

```python
plotter.get_optical(rgb=False)
```

```python
plt.imshow(plotter.optical_image.data, origin='lower')
```

```python
plotter.get_additional(surveys="WISE 3.4")
```

```python
plt.imshow(plotter.additional_image.data, origin='lower')
```

```python
plt.imshow(plotter.radio_image.data.squeeze(), origin='lower')
```

```python
plotter.plot_overlay(
    xlabel="Right Ascension [J2000]",
    ylabel='Declination [J2000]',
    sigma_list=[3.75, 50.0, 5.0],
    savefig=True
)
```
