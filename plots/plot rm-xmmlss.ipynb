```python
import astropy.units as u
import matplotlib.pyplot as plt
from astropy.coordinates import SkyCoord
from astropy.io import fits

from csromer.plots import RMPlotter
```

```python
radio_image_name = "/home/miguel/Documents/datasets/MeerKAT/MIGHTEE/XMMLSS/XMMLSS_I.fits"
rm_image = "/home/miguel/Documents/datasets/MeerKAT/MIGHTEE/XMMLSS/max_faraday_depth.fits"
pol_fraction = "/home/miguel/Documents/datasets/MeerKAT/MIGHTEE/XMMLSS/polarization_fraction.fits"
sigma_rm_image = "/home/miguel/Documents/datasets/MeerKAT/MIGHTEE/XMMLSS/sigma_phi_peak.fits"
#sigma_rm_image=None
```

```python
#center_coord = SkyCoord(ra=34.6138618 * u.deg,dec=-4.9105512 *u.deg, frame="fk5")
#center_coord = SkyCoord(ra=34.3512855 * u.deg,dec=-5.2145517 *u.deg, frame="fk5")
center_coord = SkyCoord(ra=34.2455069 * u.deg, dec=-4.8237422 * u.deg, frame="fk5")
```

```python
plotter = RMPlotter(
    rm_image=rm_image,
    rm_image_error=sigma_rm_image,
    pol_fraction_image=pol_fraction,
    total_intensity_image=radio_image_name,
    total_intensity_nsigma=5.0,
    center_coord=center_coord,
    radius=0.03 * u.deg,
    use_latex=True
)
```

```python
#plotter.plot(colorbar_label=r'RM [rad m$^{-2}$]')
plotter.plot(
    dpi=300,
    colorbar_label=[
        r'RM [rad m$^{-2}$]', r'RM Uncertainty [rad m$^{-2}$]', 'Polarization fraction'
    ],
    savefig=True
)
#plotter.plot(dpi=300, colorbar_label="Polarization fraction", savefig=True)
```
